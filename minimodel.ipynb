```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('50photos.csv')
df.head()
```

```python
df['Labels'][0]
```

```python
def get_tag(astring):
    split_list = re.split(': |,',astring)
    taglist = []
    for i in range(0,len(split_list)//4):
        taglist.append(split_list[4*i+1])
    return(taglist)
```

```python
get_tag(df['Labels'][1])
```

```python
tag_dict = {}
for i in range(len(df.index)):
    tag_dict[df['fileName'][i][20:]] = get_tag(df['Labels'][i])
tag_dict
```

```python
tag_set = set()
for key in tag_dict:
    for value in tag_dict[key]:
        tag_set.add(value)
tag_set
```
